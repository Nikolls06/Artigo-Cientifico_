## **Resumo**

Este estudo analisa a relação entre as taxas de **reprovação no Ensino Fundamental (anos finais)** e **abandono no Ensino Médio** no Brasil, no período de **2015 a 2024**.  
A hipótese central propõe que **regiões com maior reprovação no Ensino Fundamental tendem a apresentar maiores taxas de abandono no Ensino Médio**, refletindo um ciclo de exclusão escolar contínuo.

Com base em dados do **Censo Escolar (INEP)**, o trabalho realiza uma análise quantitativa exploratória utilizando **Python (pandas e matplotlib)** em ambiente **Jupyter Notebook**.  
São observadas tendências temporais, rupturas e desigualdades regionais, destacando dois marcos principais:  
(1) a **mudança metodológica de 2018**, que reduziu artificialmente as taxas de reprovação, e  
(2) os **efeitos da pandemia de COVID-19 (2020–2021)**, que provocaram queda simultânea de reprovação e abandono em todo o país.

Os resultados demonstram **correlação positiva** entre os dois fenômenos (r ≈ 0,50), indicando que **as regiões com maior reprovação no Ensino Fundamental também concentram os maiores índices de evasão no Ensino Médio**.  
Essa relação se mantém em todas as regiões brasileiras, embora com diferentes intensidades — mais forte no **Centro-Oeste (r ≈ 0,71)** e mais fraca no **Norte (r ≈ 0,43)**.

Conclui-se que a **reprovação escolar atua como um fator preditivo de abandono**, e que reduzir a evasão no Ensino Médio requer **políticas integradas de progressão e permanência** que atuem já nos anos finais do Ensino Fundamental.  
O estudo reforça a importância de compreender **a continuidade entre retenção e evasão** como parte de um mesmo desafio educacional no Brasil.

## **1. Introdução**

A trajetória escolar dos estudantes brasileiros ainda é marcada por rupturas significativas entre as etapas do ensino básico.  
Entre o final do Ensino Fundamental e o início do Ensino Médio, muitos alunos enfrentam **reprovações sucessivas**, **defasagem idade-série** e, por fim, **abandono escolar**.  
Esses fenômenos, embora distintos, fazem parte de um mesmo processo de **exclusão educacional progressiva**, que tende a se concentrar nas regiões mais vulneráveis social e economicamente.

Pesquisas nacionais (INEP, Todos Pela Educação, UNESCO) indicam que a **reprovação** é um dos principais fatores de risco para a **evasão escolar**.  
O aluno que repete o ano, especialmente nos **anos finais do Ensino Fundamental (6º ao 9º)**, acumula desmotivação, atraso e maior probabilidade de **não concluir o Ensino Médio**.  
Nesse sentido, a reprovação não representa apenas uma falha individual, mas um **indicador sistêmico de fragilidade educacional**.

Diante desse contexto, este estudo busca **investigar a relação entre a taxa de reprovação no Ensino Fundamental (anos finais)** e a **taxa de abandono no Ensino Médio**, no período de **2015 a 2024**, com base em dados agregados por **região e ano**.  
A hipótese central propõe que:

> **Regiões com maiores taxas de reprovação no Ensino Fundamental apresentam, no Ensino Médio, maiores taxas de abandono.**

Embora o efeito da reprovação sobre o abandono ocorra, em muitos casos, **com defasagem temporal de 1 a 2 anos**, o presente trabalho analisa os dados **ano a ano**, de forma simultânea, para capturar **padrões estruturais e regionais** entre os dois fenômenos.  
A opção por essa abordagem permite observar **tendências paralelas** ao longo do tempo — como quedas simultâneas, rupturas e recomposições — e **testar a consistência da hipótese** em nível nacional e regional.

A partir da análise comparativa das taxas de **reprovação e abandono**, busca-se compreender como os dois indicadores se movem em conjunto e de que forma refletem **as desigualdades regionais na trajetória escolar brasileira**.  
Mais do que uma correlação estatística, o estudo propõe uma **leitura interpretativa**, capaz de evidenciar que **onde há maior retenção, tende a haver maior evasão**, e que a permanência escolar exige políticas integradas de **progressão, aprendizagem e engajamento**.


## **2. Justificativa e Relevância do Estudo**

A educação brasileira enfrenta desafios persistentes relacionados à **retenção e à evasão escolar**, especialmente nas etapas de transição entre o Ensino Fundamental e o Ensino Médio.  
Esses dois indicadores — **reprovação** e **abandono** — não devem ser compreendidos de forma isolada, mas como **partes interdependentes de um mesmo ciclo de exclusão educacional**.

A **reprovação** nos anos finais do Ensino Fundamental frequentemente atua como um **ponto de inflexão** na trajetória do estudante: ao repetir de série, o aluno acumula defasagem idade-série, desmotivação e perda de vínculo com a escola.  
Esse processo, somado a contextos de vulnerabilidade social, contribui diretamente para o **abandono no Ensino Médio**, etapa que já exige maior engajamento e oferece menos suporte institucional.

Do ponto de vista das **políticas públicas**, compreender a relação entre reprovação e abandono é fundamental para orientar **ações preventivas**.  
Medidas que atuam exclusivamente sobre a evasão — como busca ativa de alunos ou flexibilização de matrícula — têm efeito limitado se **não enfrentarem as causas que antecedem o abandono**, como o baixo desempenho escolar, a reprovação e o descompasso entre idade e série.

Além disso, há um componente **territorial e desigual**: as regiões **Norte e Nordeste** concentram as maiores taxas de reprovação e abandono, refletindo **desigualdades históricas de investimento, infraestrutura e formação docente**.  
Portanto, investigar essas dinâmicas em escala regional permite revelar **padrões estruturais de desigualdade educacional** e reforçar a importância de políticas focalizadas e integradas.

A relevância deste estudo, portanto, está em evidenciar que **a reprovação não é apenas um indicador de desempenho escolar**, mas um **fator preditivo de exclusão futura**.  
Ao demonstrar que regiões com maior reprovação tendem a apresentar maior abandono, o trabalho busca contribuir para uma leitura **mais sistêmica e preventiva da evasão escolar**, fortalecendo o debate sobre **permanência e equidade educacional no Brasil**.


## **3. Metodologia**

### **3.1. Tipo de estudo**
Trata-se de um **estudo descritivo e exploratório**, de abordagem **quantitativa**, com foco na identificação de padrões e correlações entre dois fenômenos educacionais:  
a **reprovação no Ensino Fundamental (anos finais)** e o **abandono no Ensino Médio**, observados em diferentes **regiões do Brasil** no período de **2015 a 2024**.

O objetivo metodológico é investigar se as variações nas taxas de reprovação e abandono apresentam **movimentos paralelos ou correlacionados**, reforçando a hipótese de que **maior reprovação tende a estar associada a maior evasão**.

---

### **3.2. Fonte de dados**
Os dados utilizados foram obtidos a partir de **bases públicas oficiais** disponibilizadas pelo **Instituto Nacional de Estudos e Pesquisas Educacionais Anísio Teixeira (INEP)** e consolidadas no **Censo Escolar**.  
A base utilizada neste estudo foi estruturada em um arquivo único (`base_regioes_repro_aband_2015_2024.csv`), agregando informações anuais por **região geográfica** e por **etapa de ensino**.

As variáveis principais são:
- `ano`: período de referência (2015–2024);
- `regiao`: divisão geográfica brasileira (Norte, Nordeste, Centro-Oeste, Sudeste e Sul);
- `tx_repro_fund_af`: taxa de reprovação no Ensino Fundamental (anos finais);
- `tx_aband_med`: taxa de abandono no Ensino Médio.

---

### **3.3. Tratamento e limpeza dos dados**
Os dados foram processados no ambiente **Python (Jupyter Notebook)**, utilizando as bibliotecas **pandas** e **matplotlib**.  
As etapas incluíram:
1. Padronização dos nomes de colunas e remoção de caracteres especiais;  
2. Conversão de dados numéricos e tratamento de valores ausentes;  
3. Cálculo das **médias regionais anuais** para cada indicador;  
4. Estruturação de uma base derivada (`Região × Ano`) para análise comparativa.

---

### **3.4. Procedimentos de análise**
A análise foi dividida em três eixos complementares:

1. **Análise temporal:**  
   - Evolução das taxas médias de reprovação (EF) e abandono (EM) no Brasil (2015–2024);  
   - Identificação de rupturas em **2018** (mudança metodológica do Censo Escolar) e **2020–2021** (efeitos da pandemia).

2. **Análise regional:**  
   - Comparação entre as cinco regiões brasileiras;  
   - Visualização de desigualdades e padrões de recomposição após a pandemia.

3. **Análise relacional:**  
   - Cálculo da **correlação entre `tx_repro_fund_af` e `tx_aband_med`** (total e por região);  
   - Interpretação da dispersão e do **gap médio (EM − EF)** como medida de convergência ou distanciamento entre os dois fenômenos.

---

### **3.5. Limitações metodológicas**
Por se tratar de um estudo **agregado por ano e região**, não é possível inferir causalidade direta nem acompanhar os mesmos indivíduos ao longo do tempo.  
Além disso, as alterações nas regras de progressão (2018) e os efeitos da pandemia (2020–2021) geram **quebras na série histórica**, que foram tratadas aqui como parte da análise contextual.  
Mesmo com essas limitações, a metodologia adotada permite **compreender padrões estruturais e regionais** e **avaliar a consistência da hipótese** de associação entre reprovação e abandono.

---

**Em síntese:**  
A metodologia utilizada combina **análise temporal, regional e relacional** para identificar **tendências, rupturas e correlações** entre os dois indicadores educacionais, permitindo uma leitura integrada sobre o fluxo escolar e a permanência dos estudantes no sistema de ensino brasileiro.


```python
# 3.1 Leitura e preparação
import os, pandas as pd, numpy as np, matplotlib.pyplot as plt

CSV_PATH = r'C:\Users\taquesy\Desktop\projeto-artigo\base_regioes_repro_aband_2015_2024.csv'

df = pd.read_csv(CSV_PATH, encoding='utf-8', sep=None, engine='python')

# Padronizar colunas
df.columns = (df.columns
              .str.strip()
              .str.lower()
              .str.replace(r'[^\w\s]', '', regex=True)
              .str.replace(r'\s+', '_', regex=True))

# Coerção numérica para as métricas de interesse
for c in ['tx_repro_fund_af', 'tx_aband_med']:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

# Garantir 'ano' numérico simples (para os ticks do matplotlib)
df['ano'] = pd.to_numeric(df['ano'], errors='coerce')

# Derivados principais
brasil = (df.groupby('ano')[['tx_repro_fund_af','tx_aband_med']]
            .mean().reset_index().sort_values('ano'))
reg_ano = (df.groupby(['regiao','ano'])[['tx_repro_fund_af','tx_aband_med']]
            .mean().reset_index().sort_values(['regiao','ano']))

# Vetores de apoio p/ ticks
xticks_years = brasil['ano'].dropna().astype(int).tolist()
xticks_years_reg = reg_ano['ano'].dropna().astype(int).unique().tolist()

brasil.head(), reg_ano.head()
```

## **4. Resultados e Discussão**

A análise das taxas de **reprovação no Ensino Fundamental (anos finais)** e **abandono no Ensino Médio**, entre **2015 e 2024**, evidencia um conjunto de padrões consistentes que ajudam a compreender a dinâmica da permanência escolar no Brasil.  
As séries temporais e os indicadores regionais revelam **três momentos críticos** que marcam o comportamento das variáveis: **a estabilidade pré-2018**, **a ruptura metodológica de 2018** e **os efeitos da pandemia (2020–2021)**.

---

```python
# 4.1 Brasil — evolução das taxas (2015–2024) com destaques
plt.figure(figsize=(10,6))
plt.plot(brasil['ano'], brasil['tx_repro_fund_af'], marker='o', label='Reprovação - Fundamental (AF)')
plt.plot(brasil['ano'], brasil['tx_aband_med'],   marker='o', label='Abandono - Médio')

# Destaques (Pandemia/Metodologia)
plt.axvline(x=2018, linestyle='--', alpha=0.7)            # 2018 (mudança)
plt.axvspan(2020, 2021, alpha=0.15, label='Pandemia')     # 2020–2021

plt.title('Brasil 2015–2024 — Reprovação (EF) e Abandono (EM)')
plt.xlabel('Ano'); plt.ylabel('Taxa (%)')
plt.xticks(xticks_years); plt.legend(); plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout(); plt.show()
```


### **4.1. Tendência nacional (Figura 1)**

A Figura 1 mostra que, entre 2015 e 2017, as taxas de **reprovação no EF** se mantinham elevadas (acima de 10%), enquanto o **abandono no EM** permanecia em torno de 8%.  
Esse cenário reflete uma **trajetória de progressão escolar fragilizada**, na qual muitos alunos já enfrentavam barreiras para concluir o ensino fundamental e seguir para o ensino médio.

Em 2018, observa-se uma **queda abrupta** na reprovação — de cerca de 10% para 1% — sem redução equivalente no abandono.  
Esse comportamento coincide com **mudanças metodológicas no Censo Escolar**, especialmente no registro de alunos em progressão parcial, o que reduziu artificialmente a taxa de reprovação.  
Entre 2020 e 2021, durante a pandemia, ocorre nova distorção: tanto reprovação quanto abandono atingem **mínimos históricos** devido às políticas de **aprovação automática** e suspensão das aulas presenciais.  
A partir de 2022, ambos os indicadores voltam a subir, evidenciando a **recomposição pós-pandemia**.

Esses padrões indicam que **as duas variáveis caminham juntas ao longo do tempo**, com oscilações similares e momentos de inflexão compartilhados — sustentando a hipótese de interdependência entre reprovação e abandono.

---


```python
# 4.2 Reprovação EF por região (linhas)
regioes = reg_ano['regiao'].dropna().unique()
plt.figure(figsize=(10,6))
for r in regioes:
    sub = reg_ano[reg_ano['regiao']==r].sort_values('ano')
    plt.plot(sub['ano'], sub['tx_repro_fund_af'], marker='o', label=str(r))

plt.axvline(x=2018, linestyle='--', alpha=0.7)
plt.axvspan(2020, 2021, alpha=0.15, label='Pandemia COVID-19')
plt.title('Reprovação no Fundamental (AF) — evolução por região (2015–2024)')
plt.xlabel('Ano'); plt.ylabel('Taxa (%)')
plt.xticks(xticks_years_reg); plt.legend(); plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout(); plt.show()
```

### **4.2. Desigualdades regionais na reprovação (Figura 2)**

A evolução regional da **reprovação no EF** revela que as **regiões Norte e Nordeste** concentram historicamente os **maiores índices** do país, enquanto **Sudeste e Centro-Oeste** mantêm valores mais baixos e estáveis.  
A diferença regional é marcante: em 2017, por exemplo, o Nordeste e o Sul registravam taxas acima de 13%, contra cerca de 8% no Sudeste.  

Após 2018, todas as regiões apresentam uma queda simultânea, mas a recomposição pós-pandemia ocorre de maneira desigual: o **Norte** e o **Nordeste** voltam rapidamente a registrar níveis mais altos, evidenciando **vulnerabilidade estrutural** e **instabilidade institucional**.  
Essas variações reforçam a relação entre **reprovação e desigualdade social**, já que as regiões com menor infraestrutura educacional tendem a reproduzir padrões mais altos de retenção.

---

```python
# 4.3 Abandono EM por região (linhas)
plt.figure(figsize=(10,6))
for r in regioes:
    sub = reg_ano[reg_ano['regiao']==r].sort_values('ano')
    plt.plot(sub['ano'], sub['tx_aband_med'], marker='o', label=str(r))

plt.axvline(x=2018, linestyle='--', alpha=0.7)
plt.axvspan(2020, 2021, alpha=0.15, label='Pandemia COVID-19')
plt.title('Abandono no Ensino Médio — evolução por região (2015–2024)')
plt.xlabel('Ano'); plt.ylabel('Taxa (%)')
plt.xticks(xticks_years_reg); plt.legend(); plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout(); plt.show()
```

### **4.3. Desigualdades regionais no abandono (Figura 3)**

O comportamento do **abandono no Ensino Médio** segue trajetória semelhante.  
As regiões **Norte** e **Nordeste** apresentam as maiores taxas em todo o período (chegando a 13% e 9%, respectivamente, em 2015), enquanto **Sudeste** e **Centro-Oeste** mantêm patamares inferiores a 5%.  

Durante a pandemia, todas as regiões registram forte redução, mas a **recomposição pós-2021** é mais acentuada nas regiões de maior vulnerabilidade.  
O Norte volta a liderar o abandono escolar em 2023–2024, confirmando que **as desigualdades regionais se mantêm mesmo após eventos de ruptura**.

A semelhança entre os comportamentos das Figuras 2 e 3 sugere que **as regiões que mais reprovam também são as que mais abandonam**, sustentando a hipótese central.

---

```python
# 4.4 Dispersão EF x EM (correlação na série Região×Ano)
plt.figure(figsize=(7,5))
plt.scatter(reg_ano['tx_repro_fund_af'], reg_ano['tx_aband_med'])
plt.xlabel('Reprovação - EF (Anos Finais)')
plt.ylabel('Abandono - EM')
plt.title('Relação entre Reprovação (EF) e Abandono (EM) por Região-Ano')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout(); plt.show()

corr_total = reg_ano['tx_repro_fund_af'].corr(reg_ano['tx_aband_med'])
print(f'Correlação total (2015–2024): {corr_total:.2f}')
```

### **4.4. Relação direta entre reprovação e abandono (Figura 4)**

O gráfico de dispersão mostra uma **tendência ascendente clara** entre as duas variáveis:  
à medida que a **reprovação no EF aumenta**, cresce também o **abandono no EM**.  
A correlação geral observada é de aproximadamente **r = 0,50**, indicando uma associação moderada, porém consistente, em escala nacional.

Essa relação sugere que **a reprovação atua como um gatilho de evasão**: o aluno que reprova nos anos finais do EF tem maior probabilidade de não prosseguir no ensino médio, seja por desmotivação, defasagem ou condições socioeconômicas adversas.  
Assim, o fracasso escolar no EF se converte em **abandono futuro no EM**, caracterizando um ciclo contínuo de exclusão educacional.

---

```python
# 4.5 Força da correlação por região (barra horizontal)
corr_por_regiao = (
    reg_ano.groupby('regiao')[['tx_repro_fund_af','tx_aband_med']]
           .corr().iloc[0::2,-1]
           .reset_index()
           .rename(columns={'tx_aband_med':'correlacao'})
           .drop(columns='level_1')
)

corr_por_regiao.sort_values('correlacao', ascending=True, inplace=True)

plt.figure(figsize=(7,4))
plt.barh(corr_por_regiao['regiao'], corr_por_regiao['correlacao'])
plt.xlabel('Correlação EF x EM')
plt.title('Correlação entre Reprovação (EF) e Abandono (EM) por Região')
plt.tight_layout(); plt.show()

corr_por_regiao
```

### **4.5. Força da correlação por região (Figura 5)**

Ao observar a correlação separadamente por região, nota-se que o **Centro-Oeste (r = 0,71)** e o **Sudeste (r = 0,69)** apresentam **as associações mais fortes** entre reprovação e abandono.  
Isso indica que, nessas regiões, os movimentos de retenção e evasão são mais diretamente conectados.

Nas regiões **Nordeste (r = 0,62)** e **Sul (r = 0,58)**, a correlação também é positiva, mas com maior dispersão temporal.  
Já o **Norte (r = 0,43)** mostra uma relação mais fraca, possivelmente influenciada por variáveis externas — como migração estudantil, precariedade da infraestrutura e inconsistências administrativas.

O resultado geral reforça que, embora as intensidades variem, **todas as regiões exibem o mesmo padrão de associação positiva**, confirmando que o fenômeno é **nacional e estrutural**.

---

```python
# 4.6 Gap (EM - EF) no Brasil
gap = (brasil[['ano','tx_repro_fund_af','tx_aband_med']]
         .assign(gap=lambda x: x['tx_aband_med'] - x['tx_repro_fund_af']))

plt.figure(figsize=(8,5))
plt.plot(gap['ano'], gap['gap'], marker='o')
plt.axhline(0, linestyle='--', alpha=0.6)
plt.title('Diferença média: Abandono (EM) - Reprovação (EF) — Brasil')
plt.xlabel('Ano'); plt.ylabel('Gap (p.p.)')
plt.xticks(xticks_years); plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout(); plt.show()

gap.head()
```

### **4.6. Diferença média (gap) entre abandono e reprovação (Figura 6)**

A análise do **gap (abandono − reprovação)** complementa as evidências anteriores.  
Entre 2015 e 2017, o gap é negativo — indicando que a **reprovação superava o abandono**.  
A partir de 2018, o cenário se inverte, com um salto positivo causado pela **queda metodológica das reprovações**.  
Durante a pandemia, ambos os indicadores caem e o gap se aproxima de zero.  
Nos anos mais recentes (2022–2024), as taxas voltam a convergir, com valores próximos e estáveis.

O comportamento do gap confirma que **as variações entre reprovação e abandono estão interligadas**: quando uma variável cai por razões administrativas (como aprovação automática), a outra se torna mais visível.

## **5. Conclusão e Recomendações Finais**

Os resultados deste estudo permitem concluir que existe uma **relação estrutural e consistente entre as taxas de reprovação no Ensino Fundamental (anos finais)** e as de **abandono no Ensino Médio** no Brasil.  
A análise das séries temporais e das diferenças regionais entre 2015 e 2024 mostra que, mesmo diante de mudanças metodológicas e de eventos extraordinários como a pandemia, **as duas variáveis se movem de forma correlata**, revelando uma **interdependência sistêmica entre retenção e evasão escolar**.

De maneira geral:
- As regiões **Norte e Nordeste** apresentam as **maiores taxas** e a **maior instabilidade**, refletindo desigualdades históricas de infraestrutura, financiamento e acompanhamento escolar.  
- As regiões **Sudeste e Centro-Oeste** mantêm níveis mais baixos e estáveis, indicando **sistemas educacionais mais consolidados**.  
- A **correlação positiva** entre reprovação e abandono é observada em todas as regiões (r variando de 0,43 a 0,71), reforçando a hipótese de que **maior reprovação tende a resultar em maior evasão**.

Esses achados confirmam que a **evasão no Ensino Médio** não surge de forma repentina, mas é o **resultado acumulado de processos de exclusão que se iniciam ainda no Ensino Fundamental**.  
Cada reprovação representa não apenas um atraso acadêmico, mas também **um ponto de inflexão na motivação e na permanência do aluno**.  
Assim, reduzir o abandono no Ensino Médio exige **atuar preventivamente sobre a reprovação**, especialmente nos anos finais do EF.

---

### **5.1. Implicações para políticas públicas**
Com base nos resultados, recomenda-se que políticas de permanência e aprendizagem:
1. **Atuem nas transições entre etapas** — fortalecendo o acompanhamento dos alunos que concluem o Ensino Fundamental e ingressam no Ensino Médio.  
2. **Priorizaram as regiões Norte e Nordeste**, com estratégias específicas de combate à reprovação e programas de reforço escolar.  
3. **Monitorem continuamente os indicadores de fluxo (reprovação, aprovação e abandono)**, integrando dados do Censo Escolar e de redes estaduais.  
4. **Apoiem formações docentes e pedagógicas focadas em recuperação e engajamento**, prevenindo a reprovação precoce.  
5. **Utilizem políticas de busca ativa e recomposição de aprendizagem** não como resposta isolada à evasão, mas como parte de um sistema articulado de prevenção.

---

### **5.2. Limitações e perspectivas futuras**
Embora os resultados confirmem a hipótese em nível agregado, o estudo apresenta **limitações metodológicas**:
- Os dados são **agregados por região e ano**, não permitindo inferência individual ou causalidade direta.  
- Mudanças de metodologia (2018) e a pandemia (2020–2021) geram **quebras na série histórica**, que influenciam a comparabilidade entre anos.  
- A correlação analisada é **simultânea**; o impacto temporal da reprovação sobre o abandono poderia ser melhor investigado com **defasagem de 1 ou 2 anos**.

Para pesquisas futuras, recomenda-se:
- Aplicar **modelos de defasagem temporal (lag correlation)** para estimar o efeito da reprovação em anos anteriores sobre o abandono posterior;  
- Incorporar **dados por estado ou município**, permitindo análises mais granulares;  
- Explorar **variáveis socioeconômicas complementares** (IDEB, renda, taxa de matrícula, PIB regional) para modelar a relação entre contexto e permanência escolar.

---

### **5.3. Considerações finais**

> A hipótese proposta — de que regiões com maior reprovação no Ensino Fundamental apresentam maior abandono no Ensino Médio — foi confirmada de forma consistente.  
> A reprovação funciona como um **sinal antecipado do risco de evasão**, e a evasão representa o **desfecho visível de um processo de exclusão que começa antes da saída da escola**.

Portanto, compreender e reduzir a reprovação é, ao mesmo tempo, **prevenir o abandono e promover equidade**.  
O enfrentamento da evasão escolar no Brasil depende de **políticas articuladas entre as etapas da educação básica**, capazes de garantir **trajetórias escolares contínuas, significativas e bem-sucedidas para todos os estudantes.**
